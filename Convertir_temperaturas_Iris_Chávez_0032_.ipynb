## 1. Fórmulas de Conversión

Para transformar la temperatura entre estas dos escalas, utilizamos las siguientes expresiones matemáticas:

* **De Celsius (°C) a Fahrenheit (°F):**
    $$F = (C \times \frac{9}{5}) + 32$$
    *(También se puede usar $F = (C \times 1.8) + 32$)*

* **De Fahrenheit (°F) a Celsius (°C):**
    $$C = \frac{F - 32}{1.8}$$
    *(O su equivalente fraccionario: $C = (F - 32) \times \frac{5}{9}$)*

---

## 2. Ejemplos Resueltos Paso a Paso

### Caso A: Convertir 25°C a Fahrenheit
1.  **Identificar datos:** Tenemos $C = 25$.
2.  **Sustituir en la fórmula:** $F = (25 \times 1.8) + 32$.
3.  **Multiplicar:** $25 \times 1.8 = 45$.
4.  **Sumar:** $45 + 32 = 77$.
* **Resultado:** 25°C equivale a **77°F**.

### Caso B: Convertir 98.6°F a Celsius
1.  **Identificar datos:** Tenemos $F = 98.6$.
2.  **Sustituir en la fórmula:** $C = \frac{98.6 - 32}{1.8}$.
3.  **Restar primero (paréntesis):** $98.6 - 32 = 66.6$.
4.  **Dividir:** $66.6 / 1.8 = 37$.
* **Resultado:** 98.6°F equivale a **37°C** (temperatura corporal promedio).

## 3. Código en Python

Aquí tienes el script interactivo siguiendo las funciones y lógica solicitadas:

```python
def convertirCF():
    celsius = float(input("\nIngresa los grados Celsius: "))
    fahrenheit = (celsius * 1.8) + 32
    print(f"---> {celsius}°C es igual a {fahrenheit:.2f}°F")

def convertirFC():
    fahrenheit = float(input("\nIngresa los grados Fahrenheit: "))
    celsius = (fahrenheit - 32) / 1.8
    print(f"---> {fahrenheit}°F es igual a {celsius:.2f}°C")

def menu():
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir Celsius a Fahrenheit")
        print("2. Convertir Fahrenheit a Celsius")
        print("3. Salir")
        
        opcion = input("Selecciona una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            break
        else:
            print("Opción no válida, intenta de nuevo.")

# Llamada inicial para ejecutar el programa
if __name__ == "__main__":
    menu()
```

### Explicación breve del código:
* **`while True`**: Crea un bucle infinito para que el menú aparezca una y otra vez hasta que decidas salir.
* **`float(input(...))`**: Permite que el usuario ingrese números con decimales.
* **`f"{variable:.2f}"`**: Es un formato para que el resultado solo muestre 2 decimales y se vea más limpio.
* **`break`**: Rompe el ciclo `while` inmediatamente cuando el usuario elige la opción 3.

````python

def convertirCF():
    celsius = float(input("\nIngresa los grados Celsius: "))
    fahrenheit = (celsius * 1.8) + 32
    print(f"---> {celsius}°C es igual a {fahrenheit:.2f}°F")

def convertirFC():
    fahrenheit = float(input("\nIngresa los grados Fahrenheit: "))
    celsius = (fahrenheit - 32) / 1.8
    print(f"---> {fahrenheit}°F es igual a {celsius:.2f}°C")

def menu():
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir Celsius a Fahrenheit")
        print("2. Convertir Fahrenheit a Celsius")
        print("3. Salir")

        opcion = input("Selecciona una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            print("Iris Chavez NL = 11")
            break
        else:
            print("Opción no válida, intenta de nuevo.")

# Llamada inicial para ejecutar el programa
if __name__ == "__main__":
    menu()
````
